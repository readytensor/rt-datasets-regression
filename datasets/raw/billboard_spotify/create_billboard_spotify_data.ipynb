```python
import numpy as np
import pandas as pd
import os
```

```python
dataset_name = 'billboard_spotify'
```

```python
input_dir = './data'
inp_fname = 'music.csv'
output_dir = f'./../../processed/{dataset_name}/'
outp_fname = os.path.join(output_dir, f'{dataset_name}.csv')
```

# Read Data

```python
data = pd.read_csv(os.path.join(input_dir, inp_fname))
data.head()
```
